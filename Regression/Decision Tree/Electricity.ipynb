```python
import numpy as np
import matplotlib.pyplot as plt

# Units consumed
X = np.array([50, 120, 180, 250, 320, 380]).reshape(-1, 1)

# Corresponding Electricity Bill
y = np.array([500, 1000, 1500, 1800, 2000, 2500])
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg = DecisionTreeRegressor(max_depth=3, random_state = 42)
dt_reg.fit(X, y)
```

```python
X_grid = np.linspace(0, 400, 200).reshape(-1, 1)
y_pred = dt_reg.predict(X_grid)

plt.scatter(X, y, color = 'blue')
plt.plot(X_grid, y_pred, color = 'red')
plt.xlabel("Units consumed")
plt.ylabel("Electricity Bill ")
plt.title("Decsion Tree Regression")
plt.show()
```

```python
units = float(input("Enter electricity units consumed"))
bill = dt_reg.predict([[units]])

print(f"Predicted Electricity Bill : {bill[0]:.0f}")
```

```python
# Accuracy
accuracy = dt_reg.score(X, y) * 100
print(f"Model Accuracy: {accuracy:.2f}%")
```

Predict salary based on years of experience using Decsion Tree Regression
Input   ->  Years of Experience
Output   ->   Salary

```python
import numpy as np
import matplotlib.pyplot as plt

X1 = np.array([1, 2, 3, 4, 5, 7, 8, 9]).reshape(-1, 1)
y1 = np.array([3, 3.5, 4, 4.2, 6, 6.5, 7, 7.2])
```

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_reg1 = DecisionTreeRegressor(random_state = 42)
dt_reg1.fit(X1_train, y1_train)
```

```python
X1_grid = np.linspace(0, 8, 200).reshape(-1, 1)
y1_pred = dt_reg1.predict(X1_grid)

plt.scatter(X1, y1, color = 'blue')
plt.plot(X1_grid, y1_pred, color = 'red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary (LPA)")
plt.title("Decision Tree Regression with Train-Test Split")
plt.show()
```

```python
dt_deep = DecisionTreeRegressor(max_depth = None)
dt_deep.fit(X1_train, y1_train)

y1_pred_deep = dt_deep.predict(X1_grid)

plt.scatter(X1, y1, color = 'blue')
plt.plot(X1_grid, y1_pred_deep, color = 'green')
plt.title("OverFitting Tree")
plt.show()
```

```python
dt_controlled = DecisionTreeRegressor(max_depth = 3)
dt_controlled.fit(X1_train, y1_train)

y_pred_controlled = dt_controlled.predict(X1_grid)

plt.scatter(X1, y1, color = 'blue')
plt.plot(X1_grid, y_pred_controlled, color = 'orange')
plt.title("Controlled Depth Tree")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predictions
y1_test_pred_deep = dt_deep.predict(X1_test)
y1_test_pred_controlled = dt_controlled.predict(X1_test)

# Metrics
print("Deep Tree Metrics:")
print("MSE:", mean_squared_error(y1_test, y1_test_pred_deep))
print("R2 Score:", r2_score(y1_test, y1_test_pred_deep))

print("\nControlled Tree Metrics:")
print("MSE:", mean_squared_error(y1_test, y1_test_pred_controlled))
print("R2 Score:", r2_score(y1_test, y1_test_pred_controlled))
```
