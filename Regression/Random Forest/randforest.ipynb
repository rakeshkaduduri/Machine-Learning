```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

X = np.arange(1, 21).reshape(-1, 1)
y = np.array([
    10, 12, 15, 18, 20, 25, 30, 28, 35, 40, 
    45, 42, 50, 55, 60, 58, 65, 70, 72, 75
])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
# Decision Tree Regression 
from sklearn.tree import DecisionTreeRegressor

rf1 = DecisionTreeRegressor(random_state=42)
rf1.fit(X_train, y_train)
```

```python
# Mean Squared Error on Test Set
from sklearn.metrics import mean_squared_error

y_pred = rf.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("RandomForestRegressor Performance:")
print(f"Mean Squared Error on Test Set: {mse}")
print("R2 Score on Test Set:", rf.score(X_test, y_test))

# Visualizing the Random Forest Regression results
X_grid = np.linspace(1, 20, 200).reshape(-1, 1)
y_pred_grid = rf.predict(X_grid)
plt.scatter(X, y, color = 'blue')
plt.plot(X_grid, y_pred_grid, color = 'red')
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Random Forest Regression")
plt.show()
```

```python
# Mean Squared Error on Test Set
from sklearn.metrics import mean_squared_error

y_pred1 = rf1.predict(X_test)
mse1 = mean_squared_error(y_test, y_pred1)
print("Decision Tree Regressor Performance:")
print(f"Mean Squared Error on Test Set: {mse1}")
print("R2 Score on Test Set:", rf1.score(X_test, y_test))

# Visualizing the Random Forest Regression results
X_grid1 = np.linspace(1, 20, 200).reshape(-1, 1)
y_pred_grid1 = rf1.predict(X_grid1)
plt.scatter(X, y, color = 'blue')
plt.plot(X_grid1, y_pred_grid1, color = 'red')
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Random Forest Regression")
plt.show()
```
