### Dimensionality reduction techniques
There are two different dimensionality reduction techniques:

1. Feature Selection Methods
2. Feature Extraction Methods

--------------------------------------------------

### Feature Selection Methods

Feature Selection is the process of selecting
the most important original features and
removing irrelevant ones to improve model
performance.

- Feature Selection
    - Keeps original features
    - Just removes unnecessary ones

- Example:
    - Keep: Bedrooms, Location, Square feet
    - Remove: Wall color, Owner pet

Original columns remain unchanged.

-----------------------------------------

- Some of the common feature selection techniques are
    - Filter Methods
    - Wrapper Methods
    - Embedded Methods

-------------------------------------------------

Filter Methods:

Filter methods are a type of feature selection method that works by
selecting features based on some criteria prior to building the model.

Some of the common filtering methods are-

1. variance
2. chi-square
3. correlation coefficients
4. information gain or mutual information

---------------------------------------------------------

Variance threshold

One of the most basic filter methods is to use a variance
threshold to remove any features that have little to no
variation in their values.

Example:

If a column has:
Country = India, India, India, India
No variation -> No information -> Remove it.
Low variance = useless feature.
