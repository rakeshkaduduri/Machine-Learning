Chi-Square Test

It indicates that there is a relationship between two entities
Used for categorical data.
It checks:

Is this feature related to the target or independent?

If independent -> remove.

Example Scenario
Predict Loan Approval:
Feature: Gender
Target: Approved / Not Approved
Chi-square checks dependency.

```python
import pandas as pd
import numpy as np

X = pd.DataFrame({
    'Gender': [0,1,0,1,1],
    'Married': [1,1,0,0,1],
    'Income': [20000,40000,50000,30000,60000]
})

y = [0,1,1,0,1]

X
```

```python
# Chi-Square Requires non-negative values

from sklearn.feature_selection import chi2, SelectKBest
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

selector = SelectKBest(score_func = chi2, k = 2)
X_new = selector.fit_transform(X_scaled, y)

selected_features = X.columns[selector.get_support()]
print('Selected Features:', selected_features)
```

### Scenario: Loan Approval Classification
- You are building a classification model to predict whether a loan will be approved.
    - Dataset features:
    - Gender (Male/Female)
    - Marital Status (Yes/No)
    - Education Level (Graduate/Non-Graduate)
    - Applicant Income
    - Property Area (Urban/Rural)
    - Favorite Color
    - Loan Status (Approved/Rejected)

You want to identify which categorical features significantly influence loan approval.

- ❓ Question:
    - Which filter method is most appropriate here?
    - Which type of features can be tested using this method?
    - If “Favorite Color” has a very low chi-square score,    - what does it indicate?
    - Why is chi-square suitable only for classification problems?

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
```

```python
df = pd.DataFrame({
    "Gender": ["Male","Female","Male","Female","Male"],
    "Marital_Status": ["Yes","No","Yes","No","Yes"],
    "Education_Level": ["Graduate","Graduate","Non-Graduate","Graduate","Non-Graduate"],
    "Applicant_Income": [20000,40000,50000,30000,60000],
    "Property_Area": ["Urban","Rural","Urban","Urban","Rural"],
    "Favorite_Color": ["Red","Blue","Green","Blue","Red"],
    "Loan_Status": ["Approved","Rejected","Approved","Rejected","Approved"]
})
```

```python
le = LabelEncoder()

for col in df.columns:
    df[col] = le.fit_transform(df[col])


X = df.drop("Loan_Status", axis=1)
y = df["Loan_Status"]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
selector = SelectKBest(score_func=chi2, k=3)

X_new = selector.fit_transform(X_scaled, y)

selected_features = X.columns[selector.get_support()]

print("Selected Features:", selected_features)
```
