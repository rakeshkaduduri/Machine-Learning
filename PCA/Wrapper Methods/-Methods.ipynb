### Wrapper Methods:

Select features by actually training a machine learning
model and evaluating its performance.

- Unlike filter methods:

    - Filter -> uses statistical tests
    - Wrapper -> uses model performance

Wrapper methods wrap around a model.

That's why the name = Wrapper.

------------------------------------------------

### Types of Wrapper Methods

There are 3 major types:
1. Forward Selection
2. Backward Elimination
3. Recursive Feature Elimination (RFE)

---------------------------------------------------

### 1. Forward Selection:

- ¢ Idea
    - Start with zero features.
    - Add one feature at a time.

At each step:
- Add the feature that improves model
performance the most.

-----------------------------------------------------

### 2.Backward Elimination

- ¢ Idea
    - Start with all features.
    - Remove one feature at a time.
    - At each step:
        - Remove the least important feature.

When to Use:
- When dataset size is small
- When you want to eliminate noise gradually

------------------------------------------------

### 3. Recursive Feature Elimination (RFE)

Idea
- Instead of testing all combinations:
    1. Train model
    2. Rank features by importance
    3. Remove weakest feature
    4. Repeat

Why "Recursive"?
- Because feature elimination happens repeatedly.

-------------------------------------------------------

### Embedded Methods:

- These methods perform feature selection during the - model training
- period. This is the reason why they are called - embedded methods. Here
- the model can perform both feature selection and training at the same time.

Some of the popular methods are-
- L1 (LASSO) regularization
- L2(Ridge) regularization

decision tree based feature selection
